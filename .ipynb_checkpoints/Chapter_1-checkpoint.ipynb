````python
# æ™®é€šå‡½æ•°
def foo(x):
    print("executing foo(%s)"%(x))

class A(object):
    # å®ä¾‹æ–¹æ³•
    def foo(self,x):
        print("executing foo(%s,%s)"%(self,x))

    # ç±»æ–¹æ³•
    @classmethod
    def class_foo(cls,x):
        print("executing class_foo(%s,%s)"%(cls,x))

    # é™æ€æ–¹æ³•
    @staticmethod
    def static_foo(x):
        print("executing static_foo(%s)"%x)

a=A()
````

## 10 `*args` and `**kwargs`

ç”¨`*args`å’Œ`**kwargs`åªæ˜¯ä¸ºäº†æ–¹ä¾¿å¹¶æ²¡æœ‰å¼ºåˆ¶ä½¿ç”¨å®ƒä»¬.

å½“ä½ ä¸ç¡®å®šä½ çš„å‡½æ•°é‡Œå°†è¦ä¼ é€’å¤šå°‘å‚æ•°æ—¶ä½ å¯ä»¥ç”¨`*args`.ä¾‹å¦‚,å®ƒå¯ä»¥ä¼ é€’ä»»æ„æ•°é‡çš„å‚æ•°:

````python
def print_everything(*args):
    for count, thing in enumerate(args):
        print(f'{count}. {thing}')
print_everything('apple', 'banana', 'cabbage')
````

ç›¸ä¼¼çš„,`**kwargs`å…è®¸ä½ ä½¿ç”¨æ²¡æœ‰äº‹å…ˆå®šä¹‰çš„å‚æ•°å:

````python
def table_things(**kwargs):
    for name, value in kwargs.items():
        print(f'{name} = {value}')
table_things(apple = 'fruit', cabbage = 'vegetable')
````

ä½ ä¹Ÿå¯ä»¥æ··ç€ç”¨.å‘½åå‚æ•°é¦–å…ˆè·å¾—å‚æ•°å€¼ç„¶åæ‰€æœ‰çš„å…¶ä»–å‚æ•°éƒ½ä¼ é€’ç»™`*args`å’Œ`**kwargs`.å‘½åå‚æ•°åœ¨åˆ—è¡¨çš„æœ€å‰ç«¯.ä¾‹å¦‚:

```

```

`*args`å’Œ`**kwargs`å¯ä»¥åŒæ—¶åœ¨å‡½æ•°çš„å®šä¹‰ä¸­,ä½†æ˜¯`*args`å¿…é¡»åœ¨`**kwargs`å‰é¢.

å½“è°ƒç”¨å‡½æ•°æ—¶ä½ ä¹Ÿå¯ä»¥ç”¨`*`å’Œ`**`è¯­æ³•.ä¾‹å¦‚:


````python
def print_three_things(a, b, c):
    print(f'a = {a}, b = {b}, c = {c}')
mylist = ['aardvark', 'baboon', 'cat']
print_three_things(*mylist)
````

å°±åƒä½ çœ‹åˆ°çš„ä¸€æ ·,å®ƒå¯ä»¥ä¼ é€’åˆ—è¡¨(æˆ–è€…å…ƒç»„)çš„æ¯ä¸€é¡¹å¹¶æŠŠå®ƒä»¬è§£åŒ….æ³¨æ„å¿…é¡»ä¸å®ƒä»¬åœ¨å‡½æ•°é‡Œçš„å‚æ•°ç›¸å»åˆ.å½“ç„¶,ä½ ä¹Ÿå¯ä»¥åœ¨å‡½æ•°å®šä¹‰æˆ–è€…å‡½æ•°è°ƒç”¨æ—¶ç”¨*.

````python
# *argsï¼šæ”¶é›†å¤šä¸ªä½ç½®å‚æ•°ä¸ºå…ƒç»„
def func(*args):
    print(args)
func(1, 2, 3)
````

````python
# **kwargsï¼šæ”¶é›†å¤šä¸ªå…³é”®å­—å‚æ•°ä¸ºå­—å…¸
def func(**kwargs):
    print(kwargs)
func(a=1, b=2)
````

### 1 ä½¿ç”¨`__new__`æ–¹æ³•

````python
class Singleton(object):
    def __new__(cls, *args, **kw):
        if not hasattr(cls, '_instance'):
            orig = super(Singleton, cls)
            cls._instance = orig.__new__(cls, *args, **kw)
        return cls._instance

class MyClass(Singleton):
    a = 1

# 1. Singleton æ˜¯ä¸€ä¸ª"åªåˆ›å»ºä¸€æ¬¡å¯¹è±¡"çš„åŸºç±»ã€‚
# 2. __new__ æ˜¯æ§åˆ¶â€œåˆ›å»ºå¯¹è±¡â€çš„æ–¹æ³•ã€‚
````

### 2 å…±äº«å±æ€§

åˆ›å»ºå®ä¾‹æ—¶æŠŠæ‰€æœ‰å®ä¾‹çš„`__dict__`æŒ‡å‘åŒä¸€ä¸ªå­—å…¸,è¿™æ ·å®ƒä»¬å…·æœ‰ç›¸åŒçš„å±æ€§å’Œæ–¹æ³•.

````python
class Borg(object):
    _state = {}
    def __new__(cls, *args, **kw):
        ob = super(Borg, cls).__new__(cls, *args, **kw)
        ob.__dict__ = cls._state
        return ob

class MyClass2(Borg):
    a = 1

x = MyClass2()
y = MyClass2()
x.a = 100
print(y.a)  # 100ï¼Œè¯´æ˜çŠ¶æ€è¢«å…±äº«äº†
print(x is y)      # Falseï¼Œä¸æ˜¯åŒä¸€ä¸ªå¯¹è±¡
print(x.__dict__)  # {'a': 100}
````

### 2 æ ‡è®°-æ¸…é™¤æœºåˆ¶

åŸºæœ¬æ€è·¯æ˜¯å…ˆæŒ‰éœ€åˆ†é…ï¼Œç­‰åˆ°æ²¡æœ‰ç©ºé—²å†…å­˜çš„æ—¶å€™ä»å¯„å­˜å™¨å’Œç¨‹åºæ ˆä¸Šçš„å¼•ç”¨å‡ºå‘ï¼Œéå†ä»¥å¯¹è±¡ä¸ºèŠ‚ç‚¹ã€ä»¥å¼•ç”¨ä¸ºè¾¹æ„æˆçš„å›¾ï¼ŒæŠŠæ‰€æœ‰å¯ä»¥è®¿é—®åˆ°çš„å¯¹è±¡æ‰“ä¸Šæ ‡è®°ï¼Œç„¶åæ¸…æ‰«ä¸€éå†…å­˜ç©ºé—´ï¼ŒæŠŠæ‰€æœ‰æ²¡æ ‡è®°çš„å¯¹è±¡é‡Šæ”¾ã€‚


````python
import gc
import sys

class Node:
    def __init__(self, value):
        self.value = value
        self.children = []
        self.parent = None  # æ˜ç¡®åˆå§‹åŒ–parentå±æ€§
        print(f"åˆ›å»ºèŠ‚ç‚¹: {value}")
    
    def add_child(self, child):
        self.children.append(child)  # çˆ¶èŠ‚ç‚¹å¼•ç”¨å­èŠ‚ç‚¹
        child.parent = self           # å­èŠ‚ç‚¹å¼•ç”¨çˆ¶èŠ‚ç‚¹
        print(f"{self.value} æ·»åŠ å­èŠ‚ç‚¹ {child.value}")
    
    def __del__(self):
        print(f"é”€æ¯èŠ‚ç‚¹: {self.value}")

# è§‚å¯Ÿå¼•ç”¨è®¡æ•°å˜åŒ–
def show_refcount(obj, name):
    print(f"{name}çš„å¼•ç”¨è®¡æ•°: {sys.getrefcount(obj) - 1}")

print("=== ç¬¬ä¸€é˜¶æ®µï¼šåˆ›å»ºç‹¬ç«‹å¯¹è±¡ ===")
parent = Node("parent")
child1 = Node("child1") 
child2 = Node("child2")

show_refcount(parent, "parent")  # å¼•ç”¨è®¡æ•°: 1
show_refcount(child1, "child1")  # å¼•ç”¨è®¡æ•°: 1
show_refcount(child2, "child2")  # å¼•ç”¨è®¡æ•°: 1

print("\n=== ç¬¬äºŒé˜¶æ®µï¼šå»ºç«‹å¾ªç¯å¼•ç”¨ ===")
parent.add_child(child1)
print("å»ºç«‹parent-child1å…³ç³»å:")
show_refcount(parent, "parent")  # å¼•ç”¨è®¡æ•°: 2 (parentå˜é‡ + child1.parent)
show_refcount(child1, "child1")  # å¼•ç”¨è®¡æ•°: 2 (child1å˜é‡ + parent.children[0])

parent.add_child(child2)
print("å»ºç«‹parent-child2å…³ç³»å:")
show_refcount(parent, "parent")  # å¼•ç”¨è®¡æ•°: 3 (parentå˜é‡ + child1.parent + child2.parent)
show_refcount(child1, "child1")  # å¼•ç”¨è®¡æ•°: 2 
show_refcount(child2, "child2")  # å¼•ç”¨è®¡æ•°: 2 (child2å˜é‡ + parent.children[1])

print("\n=== ç¬¬ä¸‰é˜¶æ®µï¼šåˆ é™¤æ ˆå˜é‡ ===")
print(f"åˆ é™¤å‰åƒåœ¾å¯¹è±¡æ•°é‡: {len(gc.garbage)}")

# åˆ é™¤æ ˆä¸Šçš„å˜é‡å¼•ç”¨
del parent, child1, child2

print("æ ˆå˜é‡å·²åˆ é™¤ï¼Œä½†å¯¹è±¡ä»åœ¨å†…å­˜ä¸­ï¼ˆå› ä¸ºå¾ªç¯å¼•ç”¨ï¼‰")
print("æ²¡æœ‰çœ‹åˆ°ä»»ä½•å¯¹è±¡è¢«é”€æ¯çš„æ¶ˆæ¯")

print("\n=== ç¬¬å››é˜¶æ®µï¼šæ ‡è®°-æ¸…é™¤åƒåœ¾å›æ”¶ ===")
collected = gc.collect()
print(f"åƒåœ¾å›æ”¶å¤„ç†äº† {collected} ä¸ªå¯¹è±¡")
print("ç°åœ¨åº”è¯¥çœ‹åˆ°å¯¹è±¡è¢«é”€æ¯çš„æ¶ˆæ¯")
````
